```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
from cycler import cycler
from mpl_toolkits import mplot3d
```

```python
def density_alpha(radius):
    p0 = 1000
    p = p0*(radius/radius)
    return p
def density_delta(radius, r_star):
    p0 = 100
    frac_radius = r/r_star
    p = p0*(1-frac_radius**2)
    return p
```

```python
r = np.arange(1e-4, 6.957e10+100000, 100000)
density_d = density_delta(r, 6.957e10)
plt.plot(r, density_d, label = '$\\rho_{\delta}$')
plt.xlabel('r[cm]')
plt.ylabel('$\\rho$ [$\\frac{g}{cm^3}$]')
plt.show()
```

```python
G = 6.67e-8
k = 1.38e-16
mu = 1.3
mp = 1.67e-24

def pressure_alpha(radius, density, M_star, R_star):
    P_c = (2/3)*np.pi*G*(density)**2*R_star**2
    P_r = P_c*(1-(radius**2)/(R_star**2))
    
    return P_r

def pressure_delta(radius, central_density, R_star): 
    """
    P_c = (176/225)*np.pi*G*R_star*central_density**2
    
    P_r = P_c - ((4*np.pi*G*(central_density**2))*((radius/3)-((8*radius**3)/(45*R_star**2))+((radius**5)/(25*R_star**4))))
    """
    P_c = (4/15)*np.pi*G*(central_density**2)*(R_star**2)
    term1 = (r**2)/6
    term2 = (2*r**4)/(15*R_star**2)
    term3 = (r**6)/(30*R_star**4)
    
    P_r = P_c - (4*np.pi*G*(central_density**2)*(term1-term2+term3))
    
    return P_r
```

```python
r = np.arange(1e-4, 6.957e10+100000, 100000)
pressure_d = pressure_delta(r, 100, 6.957e10)
plt.plot(r,pressure_d)
plt.xlabel('r[cm]')
plt.ylabel('P [dyn $cm^{-2}$]')
plt.show()
```

```python
def temp_alpha(density, pressure):

    T_r = (mu*mp*pressure)/(k*density)
    
    return T_r
```

```python
temp = temp_alpha(density_d, pressure_d)

plt.plot(r,temp)
plt.xlabel('r[cm]')
plt.ylabel('T[K]')
#plt.xlim(0, 0.99e11)
plt.show()
```

```python
def rosseland_path(density, temperature):
    #gives cross section of interaction
    X = 0.7381 #solar value
    Z = 0.0134 #solar value
    g_ff = 1.0
    C = 6.25E-2
    K_r = C*(1+X)*(Z**2)*g_ff*(density)*((temperature)**(-3.5))
    return K_r

def thomson_free_electron(r):
    sigma = 0.66e-24*(r/r)
    return sigma

def free_free_opacity(density, temperature):
    C_ff = 4e22
    gff = 1
    X = 0.7381
    Y = 0.2485
    Z = 0.0134
    K_ff = C_ff*(X+Y)*(1+X)*(density/(temperature**3.5))
    return K_ff

def rosseland_mean_opacity(density, temperature):
    X = 0.7381
    Y = 0.2485
    Z = 0.0134
    
    C_ff = 3.68e18
    gff = 1
    
    K_ff = C_ff*(X+Y)*(1+X)*(density/(temperature**3.5))
    
    C_bf = 4.34e21
    g_bf = 1
    t = 1
    
    K_bf = C_bf*(g_bf/t)*Z*(1+X)*(density/(temperature**3.5))
    
    C_es = 0.02
    K_es = C_es*(1+X)
    
    if temperature <= 6000 and density <= 1e-5:
        C_H_ = 7.9e-34
        K_H_ = C_H_*(Z/0.02)*(density**-0.5)*(temperature**9)
    elif temperature >= 3000 and density >= 1e-10:
        C_H_ = 7.9e-34
        K_H_ = C_H_*(Z/0.02)*(density**-0.5)*(temperature**9)
    else:
        K_H_ = 0
    
    #K = np.mean([K_ff, K_bf, K_es]) 
    #K = ((K_ff**-1)+(K_bf**-1)+(K_es**-1))**-1
    K = K_ff+K_bf+K_es
    return K


    
```

```python
r = np.arange(1e-4, 6.957e10, 100000)
density_interaction = density_delta(r, 6.957e10)
pressure_interaction = pressure_delta(r, 100, 6.957e10)
temp_interaction = temp_alpha(density_interaction, pressure_interaction)
interaction = rosseland_path(density_interaction, temp_interaction)

plt.semilogy(r, interaction)
plt.xlabel('r[m]')
plt.ylabel('$\\sigma$ [$cm^2$]')
plt.show()

thomson_interaction = thomson_free_electron(r)

plt.plot(r, thomson_interaction)
plt.xlabel('r[m]')
plt.ylabel('$\\sigma$ [$cm^2$]')
plt.show()
```

```python
def number_density(density):
    n = density/mp
    return n
```

```python
def mean_free_path(cross_section, n_density):
    l = 1/(cross_section*n_density)
    return l
def path_probability(l, dtheta, dphi, e):
    r_theta = ((l/(1+e))*(1-e**2))/(1-e*np.cos(dtheta))
    r_phi = ((l/(1+e))*(1-e**2))/(1-e*np.cos(dphi))
    l = np.mean([r_theta, r_phi])
    #l = r_theta
    return l
def mean_free_path_beta(opacity, density):
    l = 1/(density*opacity)
    return l
```

```python
l_prob = path_probability(5, np.pi, np.pi, 0.8)
print(l_prob)
```

```python
from mpl_toolkits import mplot3d

N_steps = 100
N_layers = 10
N_walkers = 10000
c = 3e10
me = 9.1e-28

x = 0
y = 0
z = 0

x_val = [0]
y_val = [0]
z_val = [0]

r_star = 6.957e10

section = r_star/N_layers

epsilon= 1e-4

radius_bisection = np.arange(epsilon, r_star, section)
t_est = []
median_velocity = []
#fig = plt.figure(figsize = (10,10))
#ax_section = plt.axes(projection='3d')
for radius in radius_bisection:
    d_travelled = []
    t_travelled = []
    color = next(ax._get_lines.prop_cycler)['color']
    #u, v = np.meshgrid(np.arange(0,2*np.pi, 0.5), np.arange(0,np.pi,0.25))
    #sphere_x = radius * np.cos(u) * np.sin(v)
    #sphere_y = radius * np.sin(u) * np.sin(v)
    #sphere_z = radius * np.cos(v)
    #ax_section.plot_wireframe(sphere_x, sphere_y, sphere_z, alpha = 0.3, color = color)
    for walker in range(N_walkers):
        x_walker = [radius]
        y_walker = [0]
        z_walker = [0]
        x = radius
        y = 0
        z = 0
        t = 0
        r = radius
        for n in range(N_steps):
            density = density_delta(r, r_star)
            pressure = pressure_delta(r, 100, r_star)
            temperature = temp_alpha(density, pressure)
            #area_interaction = kramer_proportionality(density, temperature)
            #area_interaction = thomson_free_electron(r)
            
            opacity = rosseland_mean_opacity(density, temperature)
            
            
            theta = rand.uniform(0,2*np.pi)
            phi = rand.uniform(0,2*np.pi)
            
            l = mean_free_path_beta(opacity, density)
            #l = path_probability(l_tot, theta, phi, 0.85)
            
            
            dx = l * np.sin(phi) * np.cos(theta)
            dy = l * np.sin(phi) * np.sin(theta)
            dz = l * np.cos(phi)
           
            x = x + dx
            y = y + dy
            z = z + dz
            
            r = np.sqrt(x**2 + y**2 + z**2)

            d = r - radius

            dt = l/c
            #dt = l/electron_velocity
            t = t + dt
            
            x_val.append(x)
            y_val.append(y)
            z_val.append(z)
        
            x_walker.append(x)
            y_walker.append(y)
            z_walker.append(z)
        
        #ax_section.plot3D(x_walker,y_walker,z_walker, color = color)
        #ax_section.scatter3D(x_walker,y_walker,z_walker, color = color)
        
        d_travelled.append(d)
        t_travelled.append(t)
         

    med_d = np.mean(d_travelled)
    plt.hist(d_travelled, bins = 20)
    plt.show()
    
    med_t = np.mean(t_travelled)
    med_v = med_d/med_t
    median_velocity.append(med_v)
    
    estimated_time = section/med_v
    
    t_est.append(estimated_time)

#plt.savefig('example_photon_walk.pdf')
#fig.show()
```

```python
photon_escape_time = np.sum(np.array(t_est))
avg_photon_velocity = np.mean(np.array(median_velocity))
photon_escape_2 = r_star/avg_photon_velocity

print('estimated photon escape time = {0} [s]'.format(photon_escape_time))
print('estimated photon escape time = {0} [yr]'.format(photon_escape_time/31536000))
print('estimated photon escape time, average velocity = {0} [yr]'.format(photon_escape_2/31536000))
```

```python
N_steps = np.array([10, 100, 1000, 10000])
N_layers = 1
N_walkers = 100
c = 3e10
me = 9.1e-28

x = 0
y = 0
z = 0

x_val = [0]
y_val = [0]
z_val = [0]

r_star = 6.957e10

section = r_star/N_layers

epsilon= 1e-4

radius_bisection = np.arange(epsilon, r_star, section)
t_est = []
median_velocity = []
coordinates = []


d_travelled = []
for number_steps in N_steps:
    for walker in range(N_walkers):
        
        x_walker = [radius]
        y_walker = [0]
        z_walker = [0]
        x = radius
        y = 0
        z = 0
        t = 0
        for n in range(number_steps):

            density = density_delta(r, r_star)
            pressure = pressure_delta(r, 100, r_star)
            temperature = temp_alpha(density, pressure)
            #area_interaction = kramer_proportionality(density, temperature)
            area_interaction = thomson_free_electron(r)
            n_density = number_density(density)
        
            electron_velocity = np.sqrt(3*k*temperature/me)
        
            theta = rand.uniform(0,2*np.pi)
            phi = rand.uniform(0,2*np.pi)
            
            l_tot = mean_free_path(area_interaction, n_density)
            l = path_probability(l_tot, theta, phi, 0.99)
            
            
            dx = l * np.sin(phi) * np.cos(theta)
            dy = l * np.sin(phi) * np.sin(theta)
            dz = l * np.cos(phi)
           
            x = x + dx
            y = y + dy
            z = z + dz
            
            r = np.sqrt(x**2 + y**2 + z**2)

            d = r - radius

            #dt = l/c
            dt = l/electron_velocity
            t = t + dt
            
            x_val.append(x)
            y_val.append(y)
            z_val.append(z)
        
            x_walker.append(x)
            y_walker.append(y)
            z_walker.append(z)
        

        coordinates.append([number_steps, d])
```
