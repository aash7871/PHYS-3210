```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
from cycler import cycler
from mpl_toolkits import mplot3d
import glob as glob
```

define density profiles: density_delta used hereafter

```python
def density_alpha(radius):
    p0 = 1000
    p = p0*(radius/radius)
    return p
def density_delta(radius, r_star):
    p0 = 100
    frac_radius = r/r_star
    p = p0*(1-frac_radius**2)
    return p
```

```python
r = np.arange(1e-4, 6.957e10+100000, 100000)
density_d = density_delta(r, 6.957e10)
plt.plot(r, density_d, label = '$\\rho_{\delta}$')
plt.xlabel('r[cm]')
plt.ylabel('$\\rho$ [$\\frac{g}{cm^3}$]')
plt.show()
```

derive pressure profiles using equations of stellar structure

```python
G = 6.67e-8
k = 1.38e-16
mu = 1.3
mp = 1.67e-24

def pressure_alpha(radius, density, M_star, R_star):
    P_c = (2/3)*np.pi*G*(density)**2*R_star**2
    P_r = P_c*(1-(radius**2)/(R_star**2))
    
    return P_r

def pressure_delta(radius, central_density, R_star): 
    """
    P_c = (176/225)*np.pi*G*R_star*central_density**2
    
    P_r = P_c - ((4*np.pi*G*(central_density**2))*((radius/3)-((8*radius**3)/(45*R_star**2))+((radius**5)/(25*R_star**4))))
    """
    P_c = (4/15)*np.pi*G*(central_density**2)*(R_star**2)
    term1 = (r**2)/6
    term2 = (2*r**4)/(15*R_star**2)
    term3 = (r**6)/(30*R_star**4)
    
    P_r = P_c - (4*np.pi*G*(central_density**2)*(term1-term2+term3))
    
    return P_r
```

```python
r = np.arange(1e-4, 6.957e10+100000, 100000)
pressure_d = pressure_delta(r, 100, 6.957e10)
plt.plot(r,pressure_d)
plt.xlabel('r[cm]')
plt.ylabel('P [dyn $cm^{-2}$]')
plt.show()
```

temperature profile - derived using equation of state

```python
def temp_alpha(density, pressure):

    T_r = (mu*mp*pressure)/(k*density)
    
    return T_r
```

```python
temp = temp_alpha(density_d, pressure_d)

plt.plot(r,temp)
plt.xlabel('r[cm]')
plt.ylabel('T[K]')
#plt.xlim(0, 0.99e11)
plt.show()
```

Different methods of defining, cross section of interaction, and opacity. rosseland_mean_opacity function from Carroll and Ostile used to find mean free path. 

```python
def rosseland_path(density, temperature):
    #gives cross section of interaction
    X = 0.7381 #solar value
    Z = 0.0134 #solar value
    g_ff = 1.0
    C = 6.25E-2
    K_r = C*(1+X)*(Z**2)*g_ff*(density)*((temperature)**(-3.5))
    return K_r

def free_free_opacity(density, temperature):
    C_ff = 4e22
    gff = 1
    X = 0.7381
    Y = 0.2485
    Z = 0.0134
    K_ff = C_ff*(X+Y)*(1+X)*(density/(temperature**3.5))
    return K_ff

def rosseland_mean_opacity(density, temperature):
    X = 0.7381
    Y = 0.2485
    Z = 0.0134
    
    C_ff = 3.68e18
    gff = 1
    
    K_ff = C_ff*(X+Y)*(1+X)*(density/(temperature**3.5))
    
    C_bf = 4.34e21
    g_bf = 1
    t = 1
    
    K_bf = C_bf*(g_bf/t)*Z*(1+X)*(density/(temperature**3.5))
    
    C_es = 0.02
    K_es = C_es*(1+X)
    
    #we're able to ignore H- opacity
    """
    if temperature <= 6000 and density <= 1e-5:
        C_H_ = 7.9e-34
        K_H_ = C_H_*(Z/0.02)*(density**-0.5)*(temperature**9)
    elif temperature >= 3000 and density >= 1e-10:
        C_H_ = 7.9e-34
        K_H_ = C_H_*(Z/0.02)*(density**-0.5)*(temperature**9)
    else:
        K_H_ = 0
    
    #K = np.mean([K_ff, K_bf, K_es]) 
    #K = ((K_ff**-1)+(K_bf**-1)+(K_es**-1))**-1
    """
    K = K_ff+K_bf+K_es
    
    return K
   
```

```python
r = np.arange(1e-4, 6.957e10, 100000)
density_interaction = density_delta(r, 6.957e10)
pressure_interaction = pressure_delta(r, 100, 6.957e10)
temp_interaction = temp_alpha(density_interaction, pressure_interaction)
interaction = rosseland_path(density_interaction, temp_interaction)

plt.semilogy(r, interaction)
plt.xlabel('r[m]')
plt.ylabel('$\\sigma$ [$cm^2$]')
plt.show()
```

mean free path can de defined in different ways, (1) $l = \frac{1}{\sigma * n}$, or (2) $l = \frac{1}{\kappa * \rho}$. I chose to use definition(2).


```python
def mean_free_path(cross_section, n_density):
    l = 1/(cross_section*n_density)
    return l
def path_probability(l, dtheta, dphi, e):
    r_theta = ((l/(1+e))*(1-e**2))/(1-e*np.cos(dtheta))
    r_phi = ((l/(1+e))*(1-e**2))/(1-e*np.cos(dphi))
    l = np.mean([r_theta, r_phi])
    #l = r_theta
    return l
def mean_free_path_beta(opacity, density):
    l = 1/(density*opacity)
    return l
```

```python
all_layers =  sorted(glob.glob("rand*.txt"))
tau = []
tau_beta = []
avg_t = []
avg_d = []
avg_v = []
avg_l = []
for layer in all_layers:
    data = np.genfromtxt(layer)
    n_layer = layer.split('_')[1]
    reshaped = data.reshape(N_walkers,3)
    d_walkers = reshaped[:,0]
    t_walkers = reshaped[:,1]
    l_walkers = reshaped[:,2]
    avg_d_layer = np.mean(d_walkers)
    avg_t_layer = np.mean(t_walkers)
    avg_l_layer = np.mean(l_walkers)
    
    plt.title('distance walker traveled layer {0}'.format(n_layer))
    plt.hist(d_walkers, bins = 20)
    plt.show()
    plt.title('time walker traveled layer {0}'.format(n_layer))
    plt.hist(t_walkers, bins = 20)
    plt.show()
    plt.title('average mfp walker layer {0}'.format(n_layer))
    plt.hist(l_walkers, bins = 20)
    plt.show()
    
    avg_v_layer = avg_d_layer/avg_t_layer
    avg_d.append(avg_d_layer)
    avg_t.append(avg_t_layer)
    avg_v.append(avg_v_layer)
    avg_l.append(avg_l_layer)
    
    tau_section = section/avg_v_layer
    tau.append(tau_section)
    
    N_steps_needed = (section/avg_l_layer)**2
    tau_section_beta = N_steps_needed*avg_t_layer
    tau_beta.append(tau_section_beta)
plt.title('distance per layer')   
plt.plot(np.arange(0,10,1), avg_d, '.')
plt.xlabel('layer')
plt.ylabel('avg. distance walker travelled [cm]')
plt.savefig('rand_distance_layer.pdf')
plt.show()
plt.title('time spent per layer')
plt.plot(np.arange(0,10,1), avg_t, '.')
plt.xlabel('layer')
plt.ylabel('avg. time walker spent [s]')
plt.savefig('rand_time_layer.pdf')
plt.show()
tau = np.array(tau)
tau_beta = np.array(tau_beta)
escape_time=np.sum(tau)
escape_time_beta = np.sum(tau_beta)
```

```python
print('photon escape time = {0}[s], from random walk'.format(escape_time))
print('photon escape time = {0}[yr], from random walk'.format(escape_time/(np.pi*1e9)))
print('photon escape time = {0}[s], from random walk'.format(escape_time_beta))
print('photon escape time = {0}[yr], from number steps'.format(escape_time_beta/(np.pi*1e9)))

f = open('results_rand_walk.txt', '+w')
f.write('photon escape time = {0}[s], from random walk'.format(escape_time)+'\n'+'photon escape time = {0}[yr], from random walk'.format(escape_time/(np.pi*1e9))+'\n'+'photon escape time = {0}[s], from random walk'.format(escape_time_beta)+'\n'+'photon escape time = {0}[yr], from number steps'.format(escape_time_beta/(np.pi*1e9)))
f.close()
```
